```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
data = pd.read_csv("New Text Document.txt")
```

```python
lines = data.to_string(index=False)
```

```python
dates = []
exercises = []
weights = []

current_date = None

for line in lines.split("\n"):
    if "/" in line:
        current_date = line.split(" - ")[0].strip()
        print(line)
    elif "-" not in line:
        parts = line.split(': ')
        exercise = parts[0].strip()
        weight_str = parts[1].strip()

        weight_list = re.findall(r'[\d.]+', weight_str)
        weight_numbers = list(map(float, weight_list))

        for w in weight_numbers:
            dates.append(current_date)
            exercises.append(exercise)
            weights.append(w)
```

```python
df = pd.DataFrame({
    'Date': pd.to_datetime(dates, format='%d/%m/%Y'),
    'Exercise': exercises,
    'Weight': weights
})
```

```python
df
```

```python
plt.figure(figsize=(14, 8))

# Her egzersiz için grafik
for exercise in df['Exercise'].unique():
    subset = df[df['Exercise'] == exercise]
    plt.plot(subset['Date'], subset['Weight'], marker='o', linestyle='-', label=exercise)

plt.xlabel('Tarih')
plt.ylabel('Ağırlık (kg)')
plt.title('Egzersizlerdeki Ağırlık Değişimlerini Gösterme')
plt.legend(title='Egzersizler')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
