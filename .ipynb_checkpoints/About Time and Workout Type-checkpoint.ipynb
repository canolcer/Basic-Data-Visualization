```python
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("New Text Document.txt")
```

```python
lines = data.to_string(index=False)
```

```python
dates = []
types = []

for line in lines.split("\n"):
    if "/" in line:
        date, workout_type = line.split(' - ')
        dates.append(date.strip())
        types.append(workout_type.strip())
```

```python
dates
```

```python
df = pd.DataFrame({
    'Date': pd.to_datetime(dates, format='%d/%m/%Y'),
    'Workout Type': types
})
```

```python
workout_counts = df['Workout Type'].value_counts()
```

```python
workout_counts
```

```python
plt.figure(figsize=(10, 6))
workout_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Spor Türü')
plt.ylabel('Frekans')
plt.title('Tarihe Göre Yapılan Spor Türleri')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for workout_type in df['Workout Type'].unique():
    subset = df[df['Workout Type'] == workout_type]
    plt.plot(subset['Date'], subset['Workout Type'], marker='o', linestyle='-', label=workout_type)

plt.xlabel('Tarih')
plt.ylabel('Spor Türü')
plt.title('Tarihe Göre Yapılan Spor Türleri')
plt.legend(title='Spor Türü')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
